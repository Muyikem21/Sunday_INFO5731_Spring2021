(1) Write a Python program to find the duplicate elements in a given array of integers. Return -1 If there are no such elements. (4 points)

```python
def findDuplicates(array):
    result = []
    for x in array:
        if (array.count(x)>1) and (x not in result):
            result.append(x)
    if len(result)==0:
        return -1
    else:
        return result
print(findDuplicates([2,3,2,5,4,7]))
print(findDuplicates([1,2,3,4,5]))



```

(2) Write a Python program to select all the Sundays of a specified year. (4 points)



```python
from datetime import date, timedelta
def sundays(year):
    dt=date(year, 1, 1)
    dt += timedelta(days = 3 - dt.weekday())
    while dt.year == year:
      yield dt
      dt += timedelta(days = 7)
for s in sundays(2018):
   print(s)


```
